```python
from matplotlib import pyplot as plt
import numpy as np

for i in range(10):
    training_losses = np.loadtxt(f"./outputs/models/teacher_model_cifar_{ i + 1 }/training_losses.txt")
    accuracy = np.loadtxt(f"./outputs/models/teacher_model_cifar_{ i + 1 }/accuracy.txt")
    
    plt.plot(training_losses)
    plt.xlabel("Batch (100 intervals)")
    plt.ylabel("Loss")
    plt.title(f"Training Loss during training Teacher model #{i + 1} (accuracy: {accuracy}%)")
    plt.show()
    plt.show()

```
