```python
import matplotlib.pyplot as plt
import numpy as np

# Data - přesnosti pro různé hodnoty switch_epoch
switch_epochs = list(range(16))

accuracies_run1 = [76.45, 77.44, 77.46, 76.4, 76.38, 77.52, 77.14, 76.25, 77.1, 75.89, 76.35, 76.61, 75.5, 77.02, 75.71, 77.23]
accuracies_run2 = [76.87, 76.74, 77.1, 76.65, 76.37, 77.6, 77.12, 76.65, 76.25, 76.08, 76.43, 76.05, 76.14, 75.94, 76.31, 76.15]
accuracies_run3 = [76.64, 76.98, 76.14, 77.68, 76.6, 76.96, 77.59, 77.31, 77.18, 77.23, 76.91, 77.39, 77.23, 76.93, 75.69, 76.55]
accuracies_run4 = [76.7, 78.28, 76.65, 76.7, 76.84, 75.5, 76.64, 74.95, 76.62, 76.35, 77.5, 74.53, 76.15, 77.07, 76.05, 76.99]

# Průměrná přesnost pro každou hodnotu switch_epoch
mean_accuracies = np.mean([accuracies_run1, accuracies_run2, accuracies_run3, accuracies_run4], axis=0)

# Vykreslení jednotlivých běhů
plt.figure(figsize=(10, 6))
plt.plot(switch_epochs, accuracies_run1, marker='o', linestyle='dashed', label='Run 1', alpha=0.5)
plt.plot(switch_epochs, accuracies_run2, marker='o', linestyle='dashed', label='Run 2', alpha=0.5)
plt.plot(switch_epochs, accuracies_run3, marker='o', linestyle='dashed', label='Run 3', alpha=0.5)
plt.plot(switch_epochs, accuracies_run4, marker='o', linestyle='dashed', label='Run 4', alpha=0.5)

# Vykreslení průměrné přesnosti
plt.plot(switch_epochs, mean_accuracies, marker='o', linestyle='solid', color='black', linewidth=2, label='Mean Accuracy')

# Popisky
plt.xlabel('Switch Epoch')
plt.ylabel('Accuracy (%)')
plt.title('Impact of Switch Epoch on Model Accuracy')
plt.legend()
plt.grid()
plt.show()
```
